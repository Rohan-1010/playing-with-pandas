```python
def change(field):

    if field[0] == '>':
        
        field = field.replace('>', '')
        field = field.lstrip()
        return field
    
    elif field[0] == '<':
        
        field = field.replace('<', '')
        field = field.lstrip()
        return field
    
    elif field[0] != '>' or field[0] != '<':
        
        return field
```

```python
df['Vehicle_Age'].apply(change).reset_index().head(10)
```

```python
df['Vehicle_Age'] = df['Vehicle_Age'].apply(change)
df.head(10)
```

```python
df[['Year', 'Number']] = df['Vehicle_Age'].str.split(' ', expand = True)
```
